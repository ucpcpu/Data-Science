Importing Libraries

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Getting Datasets

```python
heart_data=pd.read_csv("/content/HeartDiseaseTrain-Test.csv")

#printing first five rows
heart_data.head()
```

```python
print(heart_data["chest_pain_type"])
print(heart_data["chest_pain_type"].value_counts())

#conversion of chest_pain_type
q=heart_data["chest_pain_type"]
w=[]
for i in q:
    if i == "Typical angina":
        w.append(0)
    elif i=="Atypical angina":
        w.append(1)
    elif i=="Non-anginal pain":
        w.append(2)
    else:
        w.append(3)

heart_data["chest_pain_type"]=w
print(heart_data["chest_pain_type"].value_counts())

print(heart_data["fasting_blood_sugar"])
print(heart_data["fasting_blood_sugar"].value_counts())

#conversion of fasting_blood_sugar
e=heart_data["fasting_blood_sugar"]
r=[]
for i in e:
    if i == "Lower than 120 mg/ml":
        r.append(1)
    else:
        r.append(0)
heart_data["fasting_blood_sugar"]=r

heart_data["fasting_blood_sugar"]=r
print(heart_data["fasting_blood_sugar"].value_counts())

print(heart_data["rest_ecg"])
print(heart_data["rest_ecg"].value_counts())

#conversion of rest_ecg
z=heart_data["rest_ecg"]
c=[]
for i in z:
    if i == "ST-T wave abnormality":
         c.append(0)
    elif i=="Normal":
         c.append(1)
    elif i=="Left ventricular hypertrophy":
         c.append(2)

heart_data["rest_ecg"]=c
print(heart_data["rest_ecg"].value_counts())

print(heart_data["slope"])
print(heart_data["slope"].value_counts())

#conversion of slope
m=heart_data["slope"]
n=[]
for i in m:
    if i == "Flat":
        n.append(1)
    elif i=="Downsloping":
        n.append(2)
    elif i=="Upsloping":
        n.append(3)
heart_data["slope"]=n

heart_data["slope"]=n
print(heart_data["slope"].value_counts())

print(heart_data["vessels_colored_by_flourosopy"])
print(heart_data["vessels_colored_by_flourosopy"].value_counts())


#conversion of vessels_colored_by_flourosopy
g=heart_data["vessels_colored_by_flourosopy"]
h=[]
for i in g:
    if i == "Zero":
        h.append(0)
    elif i=="One":
         h.append(1)
    elif i=="Two":
         h.append(2)
    elif i=="Three":
         h.append(3)
    elif i=="Four":
         h.append(4)
print(h)
heart_data["vessels_colored_by_flourosopy"]=h
print(heart_data["vessels_colored_by_flourosopy"].value_counts())

print(heart_data["thalassemia"])
print(heart_data["thalassemia"].value_counts())

#conversion of thalassemia
d=heart_data["thalassemia"]
f=[]
for i in d:
    if i == "Fixed Defect":
        f.append(1)
    elif i == "Reversable Defect":
         f.append(2)
    elif i == "Normal":
         f.append(3)
    elif i == "No":
         f.append(4)

heart_data["thalassemia"]=f
print(heart_data["thalassemia"].value_counts())

print(heart_data["exercise_induced_angina"])
print(heart_data["exercise_induced_angina"].value_counts())

#conversion of exercise_induced_angina
u=heart_data["exercise_induced_angina"]
o=[]
for i in u:
    if i == "Yes":
        o.append(0)
    else:
        o.append(1)
heart_data["exercise_induced_angina"]=o

heart_data["exercise_induced_angina"]=o
print(heart_data["exercise_induced_angina"].value_counts())

print(heart_data["sex"])
print(heart_data["sex"].value_counts())
#conversion of sex
k=heart_data["sex"]
l=[]
for i in k:
    if i == "Male":
        l.append(1)
    else:
        l.append(0)
heart_data["sex"]=l
print(heart_data["sex"])
```

Splittiing the data set

```python
X=heart_data.drop(columns="target", axis=1)
y=heart_data["target"]
print(X)
print(y)
```

```python
#train test split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, stratify=y,random_state=2)
print(X.shape,X_train.shape, X_test.shape)
print(y_test)
```

Model training using Logistic Regression

```python
model=LogisticRegression()
#training the logistic regression model with training data
model.fit(X_train, y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
print("Accuracy on training Data:", training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,y_test)
print("Accuracy on test Data:", test_data_accuracy)
```
